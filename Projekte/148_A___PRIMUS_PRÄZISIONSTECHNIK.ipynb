```python
import numpy as np
from MotorCalc import*
import matplotlib.pyplot as plt
```

```python
## Nichibo UC-280P-19160-N
```

```python
## Requirements Customer

# Dimensions (soft borders) [mm]

# Operating Voltage [V]
U_C = 13.2
# Nominal Voltage as in catalog [V]
U_N = 12
# no load speed [rpm]
n_0 = 11500
# no load current [A]
I_0 = 0.15/U_N*U_C
# stall current [A]
I_S = 2.9
# stall torque [Nm]
M_S = 250*9.81E-5
# terminal resistance [Ohm]
R = U_N/I_S
# Torque @ Working Point [Nm]
M_WP = 0.003
n_WP = 4500

motor_name_supplier = 'UC-280P-19160-N'
motor_name_KOCO = 'C2430NB0C-140-13.2-001'

# torque constant
k_M = M_S / I_S

print('terminal resistance R = {:0.2} Ohm'.format(R))
```

```python
P_mech_des=M_WP*n_WP*np.pi/30



#Print values
print('U_C = {:0.2f} V'.format(U_C))
print('I_0 = {:0.2f} A'.format(I_0))
print('k_M = {:0.2f} mNm/A'.format(k_M*1000.0))
print('W_mech = {:0.2f} W'.format(P_mech_des))
```

```python
m=CDCMotor(U_N=U_C,R=R,I_0=I_0,k_M=k_M, M_WP = M_WP, n_WP = n_WP, motor_name=motor_name_KOCO)

m.plotCurves()
```
