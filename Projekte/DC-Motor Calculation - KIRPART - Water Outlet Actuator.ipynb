```python
from MotorCalc import *
import matplotlib.pyplot as plt
import numpy as np



# gear_box GB**25R / 2430R43-14340-38 R4
GB_ratio = 4
GB_eta = 0.66


U_N=12
I_0=0.020
k_M=0.01855                #torque constant
R=10.4

M_WP = 0.05/GB_ratio*GB_eta       #torque @ Working Point
n_WP = 920*GB_ratio        #speed @ Working Point
P_mech_des=M_WP*n_WP*np.pi/30

print("desired torque before GB = {:0.2f}mNm, desired speed before GB = {:0.0f}rpm".format(1000.0*M_WP, n_WP))
print("desired mech. power = {:0.2f}W".format(P_mech_des))
```

```python
m=CDCMotor(U_N=U_N,R=R,I_0=I_0,k_M=k_M, M_WP = M_WP, n_WP = n_WP)

m.plotCurves()
```
