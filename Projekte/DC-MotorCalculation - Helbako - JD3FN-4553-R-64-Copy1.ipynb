## Constar BLDC

```python
from MotorCalc import *
import matplotlib.pyplot as plt
import numpy as np

U_N = 13.0                            #V, nominal voltage 
M_WP = 0.060                          #Nm, torque at working point
M_S = 0.180                           #Nm, stall torque
n_WP = 16000                          #rpm, speed at working point
I_S = 35                              #A, stall current
I_0 = 1                               #A, no load current
k_M = M_S/(I_S-I_0)
R = U_N / I_S                         #Ohm, terminal resistance

print('torque constant = {:0.2f}mNm/A'.format(k_M*1000.0))

```

## Variant 1 (U = 13.0V)

```python
U_N = 13.0
motor_str = 'tbd'
app_str = '147_A - Kautex Cleaning Sysmte'

m=CDCMotor(U_N=U_N,R=R,I_0=I_0,k_M=k_M,M_WP=M_WP,n_WP=n_WP,motor_name=motor_str, application=app_str)
m.plotCurves()
```

```python
P_el_WP = m.calc_P_el_from_M(M_WP)
P_mech_WP = m.calc_P_mech_from_M(M_WP)
print("torque at WP: {:0.2f}mNm".format(M_WP*1000.0))
print("calculated el. power at WP: {:0.2f}W".format(P_el_WP))
print("calculated mech. power at WP: {:0.2f}W".format(P_mech_WP))
print("loss power at WP: {:0.2f}W".format(P_el_WP-P_mech_WP))
```
