## Test Nichibo - NBV3SFN-8520F-WR-88-18928 (update 15-Nov-2018)

```python
## Import MotorCalc module
from MotorCalc import*
import matplotlib.pyplot as plt
import numpy as np


## Data Sheet Nichibo
U_N_KN = 10.8
I_S_KN = 115.415
M_S_KN = 6816*9.81E-5
I_0_KN = 2.5
motor_name_supplier = 'NBV3SFN-8520F-WR/88'
motor_name_KOCO = 'C3965NBV3SFN-174-10.8-001'
```

```python
#Terminal resistance
R=U_N_KN / I_S_KN
#Operating voltage
U=10.8
#No load current
I_0 = I_0_KN / U_N_KN * U
#Torque konstant
k_M = M_S_KN / I_S_KN
#Torque @ working point
M_WP = 0.08
#Speed @ working point
n_WP = 15000
P_mech_des=M_WP*n_WP*np.pi/30


#Print values
print('R = {:0.2f} Ohm'.format(R))
print('U = {:0.2f} V'.format(U))
print('I_0 = {:0.2f} A'.format(I_0))
print('k_M = {:0.2f} mNm/A'.format(k_M*1000.0))
print('W_mech = {:0.2f} W'.format(P_mech_des))
```

```python
m=CDCMotor(U_N=U,R=R,I_0=I_0,k_M=k_M, M_WP = M_WP, n_WP = n_WP, motor_name=motor_name_KOCO)

m.plotCurves()
```
