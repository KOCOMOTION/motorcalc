# Fischer Automotive Systems / Klappenverstellung

## Systembeschreibung
Der Kunde möchte ein bestehendes Modul zur Lüfterklappenverstellung von Johnson durch ein kleineres weniger leistungsfähiges ersetzen. Erste Tests mit einem Schrittmotor funktionieren. Auf Kundenwunsch wird noch versucht, die Systemperformance mit einem kleinen kostengünstigen DC-Motor zu erreichen.

```python
from MotorCalc import *
import matplotlib.pyplot as plt
import numpy as np

n_System_WP = 6               #Systemdrehzahl am Arbeitspunkt (rpm)
M_System_WP = 0.025           #Systemdrehmoment am Arbeitspunkt (Nm)
GB_ratio = 1000.0/6.0         #Untersetzungsverhältnis Getriebe
GB_eta = 0.6                  #angenommener Wirkungsgrad des Getriebes
app_str = '125_A / Fischer Automotive / Klappenverstellung'
```

## Motorauslegung Nichibo PC-260S-08600-N

```python
motor_str = 'PC-260S-08600-N'
n_WP = n_System_WP*GB_ratio
M_WP = M_System_WP/GB_ratio/GB_eta
R=12/0.21
U_N=8.75
k_M = (4.71)/(0.21-0.08)/1000.0
I_0 = 0.08
m=CDCMotor(U_N=U_N, k_M=k_M, R=R, I_0=I_0, M_WP=M_WP, n_WP=n_WP, motor_name=motor_str, application=app_str)
print('Motorkonstante: {:0.2f} mNm'.format(k_M*1000.0))
print('Terminal resistance: {:0.1f} Ohm'.format(R))
print('PWM On-Rate: {:0.1f} %'.format(U_N/14*100.0))
```

```python
m.plotCurves()
```

```python
P_el = m.calc_P_el_from_M(M_WP)
P_mech = m.calc_P_mech_from_M(M_WP)
P_loss = P_el-P_mech
print('elektr. Leistung P_el: {:0.2f}W'.format(P_el))
print('mech. Leistung P_mech: {:0.2f}W'.format(P_mech))
print('Verlustleistung P_loss: {:0.2f}W'.format(P_loss))
```
