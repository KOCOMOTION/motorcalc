## Motordaten

```python
import sys
sys.path.append('/Users/gerritkocherscheidt/Documents/KOCO automotive/Kompendium Motorberechnung/')
from MotorCalc import CDCMotor
import numpy

## catalog values

U_C = 12.0                    #V (as in data sheet
U_N = 6.0                     #V (applied voltage)
R = 3.05                      #Ohm (@T=-40°C)
#R = 3.85                     #Ohm (@T=20°C)
k_M = 9.9E-3                  #Nm
I_0 = 0.08/U_C*U_N            #A

## working points

n_WP = 300
M_WP = 5E-3

m=CDCMotor(U_N=U_N, I_0=I_0, k_M=k_M, R=R, H=0.61, J=6.7, n_WP=n_WP, M_WP=M_WP, \
               motor_name='BO2015NB2B03-113-12.0', application='166_A / BOSCH LiDAR')
m.calc_motor_values()
m.print_parameter()
#m.tune_voltage_to_working_point()
#m.calc_motor_values()
#m.print_parameter()
```

```python
m.plotCurves(addVoltagesSpeed=[2.0,3,4,5])
```
