```python
from MotorCalc import *
import matplotlib.pyplot as plt
import numpy as np

P_mech_des = 62.83                    #W, desired mechanical power
n_WP = 18800

M_WP = P_mech_des / n_WP *30 / np.pi  #torque, speed at working point

print("Required torque at working point {:0.2f}mNm".format(M_WP*1000))
print("Required speed at working point {:0.2f}rpm".format(n_WP))
print("Required mechanical power at working point {:0.2f}W".format(P_mech_des))
```

## Variant 1 (U = 7,2 V)

```python
R = 7.2/54.457
U_N = 7.2
I_0 = 1.8
k_M = 32.736*9.81E-5
motor_str = 'JD3FN-5521SF-R/60'
motor_str_KOCO = "C3047N0BC-210-7.2-001"
app_str = 'Helbako - Positioniereinheit, V=7,2V'

m=CDCMotor(U_N=U_N,R=R,I_0=I_0,k_M=k_M,M_WP=M_WP,n_WP=n_WP,motor_name=motor_str_KOCO, application=app_str)
m.plotCurves()
M_WP = m.calc_M_from_n(n_WP)
```

```python
print(m.calc_n_from_M(0.017))
```

```python
P_el_WP = m.calc_P_el_from_M(M_WP)
P_mech_WP = m.calc_P_mech_from_M(M_WP)
print("torque at WP: {:0.2f}mNm".format(M_WP*1000.0))
print("calculated el. power at WP: {:0.2f}W".format(P_el_WP))
print("calculated mech. power at WP: {:0.2f}W".format(P_mech_WP))
print("loss power at WP: {:0.2f}W".format(P_el_WP-P_mech_WP))
```
