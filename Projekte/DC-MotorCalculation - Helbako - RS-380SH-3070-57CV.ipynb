## RS-380-SH-3070-57CV

```python
from MotorCalc import *
import matplotlib.pyplot as plt
import numpy as np

motor_name='RS-380-SH-3070-57CV'

P_mech_des = 62.83                    #W, desired mechanical power

M_WP = 0.06                           #Nm, torque at working point
n_WP = P_mech_des / M_WP *30 / np.pi  #rpm, speed at working point

print("Required torque at working point {:0.2f}mNm".format(M_WP*1000))
print("Required speed at working point {:0.2f}rpm".format(n_WP))
print("Required mechanical power at working point {:0.2f}W".format(P_mech_des))
```

## Variant 1 (U = 7,2 V)

```python
R = 9/8.5
U_N = 7.2
I_0 = 0.280
k_M = 549/8.5*9.81E-5

m=CDCMotor(U_N=U_N,R=R,I_0=I_0,k_M=k_M,M_WP=M_WP,n_WP=n_WP,motor_name=motor_name)
M_WP = m.calc_M_from_n(n_WP)
m=CDCMotor(U_N=U_N,R=R,I_0=I_0,k_M=k_M,M_WP=M_WP,n_WP=n_WP,motor_name=motor_name)
m.plotCurves()
```
